```python
import unicodecsv
with open('mefile.csv','wb') as f:
    f.write(u'\ufeff'.encode('utf8'))
    w = unicodecsv.writer(f, dialect=unicodecsv.excel, delimiter=";", quoting=unicodecsv.QUOTE_ALL, encoding='utf8')
    w.writerow((u'Δημήτρης', u'Κώτσιρας', u'“Μεταπτυχιακό στην Επιστήμη Δεδομένων”'))
```
